```python
dictionary={}
for i,word in enumerate(book_txt):
    if word.lower() in possible_words:
        if word.lower() in list(dictionary.keys()):
            dictionary[word.lower()]+=1
        else:
            dictionary[word.lower()]=1 
    if i%1000==0:
        print(i)
```

```python
output= ['B', 'B','B' ,'Y','B']
```

```python
def output_possible_letters(guess, output, p_lk=possible_letters):
    p_l=p_lk.copy()
    guess=guess.lower()
    guess=list(guess)
    for i,col in enumerate(output):
        if col=='G':
            if p_l[guess[i]][0] != 'present':
                p_l[guess[i]][0]='present'
            p_l[guess[i]].append(['G',i])
        elif col=='Y':
            if p_l[guess[i]][0] != 'present':
                p_l[guess[i]][0]='present'
            p_l[guess[i]].append(['Y',i])
        else:
            if p_l[guess[i]][0] == 'present':
                p_l[guess[i]].append(['Y',i])
            else:
                p_l[guess[i]][0]='absent'
    return p_l
```

```python
b=output_possible_letters(guess, output, possible_letters)
yes_sir_list=[(k,v[1:]) for (k,v) in b.items() if v[0]=='present']
absent_list= [(k,v[0]) for (k,v) in b.items() if v[0]=='absent']
position_list=[]
present_list= []
for (k,lis) in yes_sir_list:
    num=len(lis)
    for i in range(num):
        if lis[i][0]== 'Y':
            present_list.append((k,lis[i][1]))
        else:
            position_list.append((k,lis[i][1]))
```

```python
print(guess_df.shape[0])
for l,p in position_list:
    guess_df= guess_df.loc[guess_df[str(p)]==l]
for l,q in present_list:
    guess_df= guess_df.loc[~(guess_df[str(q)]==l)]
    guess_df= guess_df.loc[(guess_df.loc[:,:]==l).sum(axis=1).astype('bool')]
for l,_ in absent_list:
    guess_df=guess_df.loc[~(guess_df.loc[:,:]==l).sum(axis=1).astype('bool')]
if guess_df['word'].to_list()[0]==guess:
    guess_df=guess_df.iloc[1:]
guess_list=guess_df['word'].to_list()
score=0
guess='nothing'
for guesses in guess_list:
    if guesses in better_words_lst:
        likely_score=dictionary[guesses]
        if likely_score > score:
            score=likely_score
            guess=guesses
            print(guess,score)
        elif likely_score > 0.05*score:
            print(guesses,likely_score)
print(guess_df.shape[0])
```
