```python
def output_possible_letters(guess, output, p_lk=possible_letters):
    p_l=p_lk.copy()
    guess=guess.lower()
    guess=list(guess)
    for i,col in enumerate(output):
        if col=='G':
            if p_l[guess[i]][0] != 'present':
                p_l[guess[i]][0]='present'
            p_l[guess[i]].append(['G',i])
        elif col=='Y':
            if p_l[guess[i]][0] != 'present':
                p_l[guess[i]][0]='present'
            p_l[guess[i]].append(['Y',i])
        else:
            if p_l[guess[i]][0] == 'present':
                p_l[guess[i]].append(['Y',i])
            else:
                p_l[guess[i]][0]='absent'
    return p_l
```

```python
def probability_calculation(word, combination, word_list):
    guess_df=[list(words) for words in word_list]
    guess_df=pd.DataFrame(guess_df)
    guess_df.columns=['0','1','2','3','4']
    guess_df['word']= word_list
    p_l={
    'a': [-1],
    'b': [-1],
    'c': [-1],
    'd': [-1],
    'e': [-1],
    'f': [-1],
    'g': [-1],
    'h': [-1],
    'i': [-1],
    'j': [-1],
    'k': [-1],
    'l': [-1],
    'm': [-1],
    'n': [-1],
    'o': [-1],
    'p': [-1],
    'q': [-1],
    'r': [-1],
    's': [-1],
    't': [-1],
    'u': [-1],
    'v': [-1],
    'w': [-1],
    'x': [-1],
    'y': [-1],
    'z': [-1],
    }
    word=word.lower()
    guess=list(word)
    for i,col in enumerate(combination):
        if col=='G':
            if p_l[guess[i]][0] != 'present':
                p_l[guess[i]][0]='present'
            p_l[guess[i]].append(['G',i])
        elif col=='Y':
            if p_l[guess[i]][0] != 'present':
                p_l[guess[i]][0]='present'
            p_l[guess[i]].append(['Y',i])
        else:
            if p_l[guess[i]][0] == 'present':
                p_l[guess[i]].append(['Y',i])
            else:
                p_l[guess[i]][0]='absent'
    yes_sir_list=[(k,v[1:]) for (k,v) in p_l.items() if v[0]=='present']
    absent_list= [(k,v[0]) for (k,v) in p_l.items() if v[0]=='absent']
    position_list=[]
    present_list= []
    for (k,lis) in yes_sir_list:
        num=len(lis)
        for i in range(num):
            if lis[i][0]== 'Y':
                present_list.append((k,lis[i][1]))
            else:
                position_list.append((k,lis[i][1]))
    for l,p in position_list:
        guess_df= guess_df.loc[guess_df[str(p)]==l]
    for l,q in present_list:
        guess_df= guess_df.loc[~(guess_df[str(q)]==l)]
        guess_df= guess_df.loc[(guess_df.loc[:,:]==l).sum(axis=1).astype('bool')]
    for l,_ in absent_list:
        guess_df=guess_df.loc[~(guess_df.loc[:,:]==l).sum(axis=1).astype('bool')]
    probability=guess_df.shape[0]/len(word_list)  
    #info= probability*np.log2(1/probability) if probability != 0 else 0
    return (probability,guess_df)
```

```python
from joblib import Parallel, delayed
import multiprocessing
inputs = remain_lst
possiblity= possibility #or remain_lst
def info_calculation(word,all_possible_combination=all_possible_combination, words=inputs):
    info=0
    hp=0
    shp=0
    word_dict={}
    for combination in all_possible_combination:
        z,_ = probability_calculation(word,combination,words)
        s=""
        combination=s.join(combination)
        word_dict[combination]=z
        infoi = z*np.log2(1/z) if z != 0 else 0
        info +=infoi
        if z > 0.15:
            hp += 1
        if z > 0.25:
            shp += 1
    #if hp >=2 or shp >=1:
        #info = 0
    word_dict['information']=info
    return word_dict #info
num_cores = multiprocessing.cpu_count()
results = Parallel(n_jobs=num_cores)(delayed(info_calculation)(i) for i in possibility) #inputs to be changed by possibility
```
