```python
def guess_check(guess, word, possible_letters, guess_lst=guess_lst):
    if guess in guess_lst:
        guess=guess.lower()
        word=word.lower()
        result=[]
        guess=list(guess)
        word =list(word)
        for i,l in enumerate(guess):
            if l not in word:
                result.append('B')
                #possible_letters[l]= [-999,i]
            else:
                if l==word[i]:
                    result.append('G')
                    #possible_letters[l]= [i,i]
                else:
                    result.append('Y')
                    #possible_letters[l]= [0,i] 
        return result, possible_letters
    else: 
        print ('guess not a word')
```
